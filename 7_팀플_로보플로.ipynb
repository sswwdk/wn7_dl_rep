# 하이퍼 파라미터 튜닝 후 1

```python
# 하이퍼 파라미터 튜닝 후 훈련에 적용 1

from ultralytics import YOLO
import yaml

# YOLOv8 모델 로드
model = YOLO("yolov8n.pt")

# 데이터셋 경로 설정
data_path = "./datasets2/data.yaml"

# 하이퍼파라미터 yaml 파일 불러오기
with open("runs/detect/tune4/best_hyperparameters.yaml", "r", encoding="utf-8") as f:
    hyp_dict = yaml.safe_load(f)

# 모델 훈련
model.train(data=data_path, epochs=10, imgsz=640, **hyp_dict)
```
